# Обучим рандомный лес

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
from sklearn.metrics import accuracy_score
```

```python
clf = RandomForestClassifier()
params = {
    'max_depth':range(3,5,1)
}
```

```python
grid_search_rf = GridSearchCV(clf, params)
```

```python
%%time
grid_search_rf.fit(X_train, y_train)
```

```python
best_rf = grid_search_rf.best_estimator_
y_pred = best_rf.predict(X_test)
```

```python
cross_val_score(best_rf, X_train, y_train, cv=5, scoring="accuracy").mean()
```

```python
y_pred = pd.DataFrame(y_pred, columns=['Survived']).reset_index()
first_y = gender_submission.drop(columns='Survived').reset_index()
answer_forest = pd.merge(first_y, y_pred,  how='inner', left_on='index', right_on='index')
answer_forest = answer_forest.drop(columns='index')
answer_forest.head()
```

```python
answer_forest.to_csv('rf.csv', index=False)
```

```python
# score 0.76
```

# Настроим xgboost в два этапа

На первом этапе подберем с помощью RandomizedSearchCV коэффициенты регуляризации и настроим сложность модели.

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBClassifier(num_class=2)
```

```python
params = {'n_estimators':[20],
          'eta':[0.001],
        'max_depth': np.arange(3,14,1),
        'min_child_weight': np.arange(1,10,1),
        'subsample': np.arange(0.5,1,0.05),
        'gamma': np.arange(0.5,1,0.01),
        'colsample_bytree': np.arange(0.4, 1, 0.05),
        'objective':['multi:softprob'],
          'eval_metric':['merror']
    }
```

```python
rand_search_Xgb = RandomizedSearchCV(xgb_model, params, cv=5, n_iter=400)
```

```python
%%time
rand_search_Xgb.fit(X_train,y_train)
```

```python
best_params = rand_search_Xgb.best_params_
best_params
```

```python
best_params.pop('eta')
```

```python
best_params['num_class'] = 2
best_params
```

На втором этапе подберем лучший learning rate и n_estimators

```python
xgb_model2 = xgb.XGBClassifier(n_estimators=20,objective='multi:softprob',subsample=0.95,min_child_weight=1,max_depth=9,gamma=0.54,
                              eval_metric='merror',colsample_bytree=0.80,num_class=2)
```

```python
params2 = {
    'eta':np.arange(0.005, 0.05, 0.005)
}
```

```python
grid_best_eta = GridSearchCV(xgb_model2, params2, cv=5)
```

```python
grid_best_eta.fit(X_train, y_train)
```

```python
grid_best_eta.best_params_
```

```python
best_params.pop('n_estimators')
```

```python
best_params['eta'] = 0.04
best_params
```

```python
dtrain = xgb.DMatrix(X_train, y_train)
```

```python
%%time
xgbCVResult = xgb.cv(best_params, dtrain, num_boost_round=500, nfold=5, early_stopping_rounds=100)
```

```python
xgbCVResult.shape
```

```python
xgbCVResult.head()
```

```python
plt.plot(range(xgbCVResult.shape[0]), xgbCVResult['test-merror-mean'], label='test-merror-mean')
plt.plot(range(xgbCVResult.shape[0]), xgbCVResult['train-merror-mean'], label='train-merror-mean')
plt.legend()
```

```python
best_num_round = np.argmin(xgbCVResult['test-merror-mean'])
best_num_round
```

```python
bestXgb = xgb.train(best_params, dtrain, num_boost_round=best_num_round)
```

```python
y_test = gender_submission.set_index('PassengerId')
```

```python
dtest = xgb.DMatrix(X_test, y_test)
```

```python
xgb_predict = bestXgb.predict(dtest)
```

```python
xgb_prediction = np.argmax(xgb_predict, axis=1)
xgb_prediction
```

```python
y_pred = pd.DataFrame(xgb_prediction, columns=['Survived']).reset_index()
first_y = gender_submission.drop(columns='Survived').reset_index()
answer_xgb = pd.merge(first_y, y_pred,  how='inner', left_on='index', right_on='index')
answer_xgb = answer_xgb.drop(columns='index')
answer_xgb.head()
```

```python
answer_xgb.to_csv('xgb.csv', index=False)
```
